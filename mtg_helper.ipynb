```python
import pandas as pd
from matplotlib import pyplot as plt
from urllib.request import urlretrieve, urlopen
import json
import pandas as pd
from matplotlib import pyplot as plt
import os,sys,time
```

```python
time.sleep(1)
```

```python
try:
  mwCards = pd.read_csv('./mwCards.csv')
  print('found csv!')
except:
  print('no csv!  loading from scryfall!')  
  bulkfiles_json = urlopen('https://api.scryfall.com/bulk-data').read()
  file_dict = {r['description']:r['download_uri'] for r in json.loads(bulkfiles_json)['data']}
  oracle_urls = [d for k,d in file_dict.items() if 'Oracle' in k]
  time.sleep(3)
  oracle_cards_json = urlopen(oracle_urls[0]).read()
  oracle_cards = pd.DataFrame(json.loads(oracle_cards_json))
  mwCards = oracle_cards[[not any([x in str(mc) for x in 'RBGP']) for mc in oracle_cards.mana_cost]]
  mwCards.to_csv('./mwCards.csv', index=False)
```

```python
def find_all(cards, col, words):
  return(
      cards[[all([w in str(txt).lower() for w in words]) for txt in cards[col]]]
  )
find_all(mwCards, 'oracle_text', ['gain life', 'whenever'])
```
